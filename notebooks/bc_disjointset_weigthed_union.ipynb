## Quick union

```python
class DisjointSet:
  def __init__(self, n):
    self.s = [-1]*n

  def find(self, e):
    if self.s[e] < 0:
      return e
    else:
      return self.find(self.s[e])

  def union(self, a, b):
    A = self.find(a)
    B = self.find(b)
    if A != B:
      if abs(self.s[B]) > abs(self.s[A]):
        self.s[B] += self.s[A]
        self.s[A] = B
      else:
        self.s[A] += self.s[B]
        self.s[B] = A
```

```python
ds = DisjointSet(8)
print(ds.s)
print(ds.find(5))
print(ds.find(7))
ds2gv(ds)
```

```python
ds.union(5, 7)
print(ds.s)
print(ds.find(5))
print(ds.find(7))
ds2gv(ds)
```
