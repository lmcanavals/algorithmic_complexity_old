```python
import networkx as nx
import math
```

```python
def relax(G, u, v):
  g_u = G.nodes[u]["dist"]
  g_v = G.nodes[v]["dist"]
  w_uv = G.edges[u, v]["weight"]
  if g_u + w_uv < g_v:
    G.nodes[v]["dist"] = g_u + w_uv
    G.nodes[v]["path"] = u
    return True
  return False

def bellmanFord(G, s):
  n = G.number_of_nodes()
  for u in G.nodes:
    G.nodes[u]["dist"] = math.inf
    G.nodes[u]["path"] = -1

  G.nodes[s]["dist"] = 0

  for _ in range(n - 1):
    for u, v in G.edges:
      relax(G, u, v)

  for u, v in G.edges:
    if relax(G, u, v):
      return False
  return True
```
